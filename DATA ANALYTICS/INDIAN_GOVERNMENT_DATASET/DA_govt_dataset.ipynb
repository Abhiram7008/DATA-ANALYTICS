#  Correlation Heatmap

```python
corr = df[['Min_x0020_Price', 'Max_x0020_Price', 'Modal_x0020_Price']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap of Prices')
plt.show()
```
